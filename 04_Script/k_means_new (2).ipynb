导入数据+标准化

```python
import pandas as pd
data = pd.read_csv('Fishnet_Join_FNL__.csv')
df = pd.DataFrame(data)
```

```python
df
```

```python
df = df[["OID_", "Medical_Num", "NEAR_DIST_Medic", "NEAR_DIST_Senior", "NEAR_DIST_Trans", "NEAR_DIST_Subw","NEAR_DIST_sch_1","NEAR_DIST_Plaza_1",
        "DIST_Culture","SchoolNum","Pct_Resd","Pct_Comer"]]
df
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn import preprocessing


#df.convert_objects(convert_numeric=True)#将object格式转float64格式
df.fillna(0, inplace=True)#把NaN替换为0

print(df.head())
df = df.set_index('OID_')
# 由于是非监督学习，不使用label
x = np.array(df).astype(float)
# 将每一列特征标准化为标准正态分布，注意，标准化是针对每一列而言的
x = preprocessing.scale(x) #(X-X_mean)/X_std,最后得到的结果是，对每个属性/每列来说所有数据都聚集在0附近，方差值为1
#x
```

```python
df
```

```python
x
```

```python
estimator = KMeans(n_clusters=6) # 构造聚类器
estimator.fit(x) # 聚类
label_pred = estimator.labels_ # 获取聚类标签
```

```python
label_pred
```

```python
#用Calinski-Harabasz Index评估的聚类分数:
from sklearn import metrics

m6 = metrics.calinski_harabaz_score(x,label_pred)  
```

通过改变n_clusters数值，取Calinski-Harabasz Index评估的聚类分数最高时的n_clusters

```python
estimator3 = KMeans(n_clusters=3) # 构造聚类器
estimator3.fit(x) # 聚类
label_pred3 = estimator3.labels_ # 获取聚类标签、
m3 = metrics.calinski_harabaz_score(x,label_pred3)
```

```python
estimator4 = KMeans(n_clusters=4) # 构造聚类器
estimator4.fit(x) # 聚类
label_pred4 = estimator4.labels_ # 获取聚类标签、
m4 = metrics.calinski_harabaz_score(x,label_pred4)
```

```python
estimator5 = KMeans(n_clusters=5) # 构造聚类器
estimator5.fit(x) # 聚类
label_pred5 = estimator5.labels_ # 获取聚类标签、
m5 = metrics.calinski_harabaz_score(x,label_pred5)
```

```python
estimator7 = KMeans(n_clusters=7) # 构造聚类器
estimator7.fit(x) # 聚类
label_pred7 = estimator7.labels_ # 获取聚类标签、
m7 = metrics.calinski_harabaz_score(x,label_pred7)
```

```python
estimator8 = KMeans(n_clusters=8) # 构造聚类器
estimator8.fit(x) # 聚类
label_pred8 = estimator8.labels_ # 获取聚类标签、
m8 = metrics.calinski_harabaz_score(x,label_pred8)
```

```python
import matplotlib.pyplot as plt
x1 = [3, 4, 5, 6, 7, 8]
y1 = [m3,m4,m5,m6,m7,m8]
plt.plot(x1, y1, "g", marker='D', markersize=5, label="Calinski-Harabasz Index")
plt.xlabel("clusters")
plt.ylabel("score")
plt.legend(loc="lower right")
plt.show()
```

```python
#重新聚类为10
estimator = KMeans(n_clusters=10) # 构造聚类器
estimator.fit(x) # 聚类
label_pred = estimator.labels_ # 获取聚类标签

label_pred
```

```python
df['label'] = label_pred.tolist()
df
```

```python
df.to_csv("data_with_label_10.csv")
```
