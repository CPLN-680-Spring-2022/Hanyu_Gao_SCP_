```python
import glob
import pandas as pd
import geopandas as gpd
```

POI data source: https://opendata.pku.edu.cn/dataset.xhtml?persistentId=doi:10.18170/DVN/WSXCNM

```python
files = glob.glob('/content/drive/MyDrive/Sponge City/POI/CSV/*.csv')
```

```python
list_of_df = list()

for i in range(0, len(files)):
  df = pd.read_csv(files[i], encoding='GBK', engine='python', error_bad_lines=False)

  df = df.loc[df['cityname'] == '郑州市']

  list_of_df.append(df)
```

```python
type(list_of_df)
```

```python
df = pd.concat(list_of_df, ignore_index=True)
```

```python
list(df.columns)
```

```python
len(df)
```

```python
df.head(1)
```

```python
df[['X','Y']] = df['location'].str.split('，',expand=True)
```

```python
df[['type_1', 'type_2', 'type_3', 'type_4', 'type_5', 
       'type_6', 'type_7', 'type_8', 'type_9']] = (
           df['type'].str.split(';', expand=True).rename(columns=lambda x: f"type_{x+1}")
           )
```

```python
ZZpoi_df = df[['name', 'cityname', 'adname', 'address', 'X', 'Y', 'type_1', 'type_2', 'type_3']].reset_index()
```

```python
ZZpoi_df.head()
```

```python
print(ZZpoi_df['type_1'].value_counts())
```

```python
ZZpoi_gdf = gpd.GeoDataFrame(ZZpoi_df, geometry=gpd.points_from_xy(ZZpoi_df.X, ZZpoi_df.Y))
```

```python
len(ZZpoi_gdf)
```

```python
ZZpoi_gdf.to_file('/content/drive/MyDrive/Sponge City/POI/ZZpoi.shp', encoding='GB18030')  
```

```python
ZZpoi_gdf.to_file('/content/drive/MyDrive/Sponge City/POI/ZZpoi.geojson', driver='GeoJSON')  
```
